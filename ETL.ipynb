```python
customers=pd.read_json("customers.json")
```

```python
customers_csv_path="customers.csv"
```

```python
customers.csv=pd.read_csv(customers_csv_path)
```

```python
display(customers.csv)
```

```python
#merge the 2 csv data
```

```python
order_details=pd.merge( customers.csv,orders, on='Order ID')
```

```python
display(order_details)
```

```python
#explore the data
```

```python
order_details.describe()
```

```python
order_details.info()
```

```python
#handling missing values
```

```python
missing_values=order_details.isnull().sum()
```

```python
display("missing_values:\n", missing_values)
```

```python
order_details=order_details.dropna()
```

```python
#handling duplicates
```

```python
order_details.to_csv("order_details.csv" , index=False)
```

```python
has_duplicates = order_details.duplicated(subset='Order ID').any()
```

```python
if has_duplicates:
    print("There are duplicate rows based on the 'Order ID' column.")
else:
    print("There are no duplicate rows based on the 'Order ID' column.")
```

```python
order_details = order_details.drop_duplicates(subset='Order ID')
```

```python
display(order_details)
```

```python
#rename the columns
```

```python
order_details.rename(columns={'Amount': 'TotalPrice'}, inplace=True)
```

```python
order_details.rename(columns={'Quantity': 'UnitsOrdered'}, inplace=True)
```

```python
order_details.rename(columns={'Sub-Category': 'Product'}, inplace=True)
```

```python
display(order_details)
```

```python
data_types = order_details.dtypes
```

```python
display(data_types)
```

```python
#Convert 'Order Date' to datetime and extract the month
```

```python
order_details['Order Date'] = pd.to_datetime(order_details['Order Date'] , format='mixed')
```

```python
order_details["Month"] = order_details["Order Date"].dt.month
```

```python
display(order_details)
```

```python
order_details.to_csv("order_details.csv" , index=False)
```

#  PerformanceMetric

```python
#category_based aggregation
```

```python
sorted_category_sales = order_details.groupby('Category')['UnitsOrdered'].sum().reset_index().sort_values(by='UnitsOrdered', ascending=False)
display("Best Performing Categories:")
display(sorted_category_sales )
```

```python
sorted_category_sales.to_csv("sorted_category_sales.csv" , index=False)
```

```python
#product_based aggregation
```

```python
sorted_product_sales = order_details.groupby(['Category', 'Product'])['UnitsOrdered'].sum().reset_index().sort_values(by='UnitsOrdered', ascending=False)
```

```python
sorted_product_sales.to_csv("sorted_product_sales", index=False)
```

```python
display("Performance of Different Categories and Products by Number of Units Sold:")
display(sorted_product_sales)
```

# ProfitabilityMetric

```python
order_details['ProfitMargin'] = (order_details['Profit'] / order_details['TotalPrice']) * 100
```

```python
product_profit_margin = order_details.groupby('Product')['ProfitMargin'].mean().reset_index().sort_values(by='ProfitMargin', ascending=False)
```

```python
product_profit_margin.to_csv("product_profit_margin .csv", index=False)
```

```python
display(product_profit_margin)
```

```python
order_details.to_csv("order_details.csv" , index=False)
```

# Geographical aggregation

```python
sorted_sales_per_city= order_details.groupby(['State', 'City'])['TotalPrice'].sum().reset_index().sort_values(by='TotalPrice', ascending=False)
display(sorted_sales_per_city)
```

```python
sorted_sales_per_city.to_csv("sorted_sales_per_city.csv")
```

# Seasonal pattern

```python
monthly_sales= order_details.groupby('Month')['TotalPrice'].sum().reset_index().sort_values(by='TotalPrice', ascending=False)
```

```python
month_mapping = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
                 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
```

```python
monthly_sales['Month'] = monthly_sales['Month'].map(month_mapping)
```

```python
display(monthly_sales)
```

```python
monthly_sales.to_csv("monthly_sales.csv" , index=False)
```

```python
order_details['CustomerStatus'] = order_details.duplicated(subset='CustomerName', keep='first').map({True: 'Repeat', False: 'New'})
```
